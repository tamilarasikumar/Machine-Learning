```python

import matplotlib.pyplot as plt
%matplotlib inline

days=  [1,2,3,4,5,6,7]
max_t= [50,51,52,48,49,46,43]
min_t= [43,42,40,44,33,35,37]
avg_t= [45,48,48,46,40,42,41]

plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.legend(loc='best',shadow=True)
plt.grid()


                   
           
```

```python
#bar chart in matplotlib

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

company=['googl','amzn','fb','msft']
revenue=[90,135,89,35]
profit=[40,2,34,12]

xpos=np.arange(len(company))
```

```python
plt.xticks(xpos,company)

plt.ylabel("revenue (bill)")
plt.title('us techstock')

#plt.bar(xpos,revenue,label="revenue")
#plt.bar(xpos,profit,label="profit")

plt.bar(xpos-0.2,revenue,label="revenue")
plt.bar(xpos+0.2,profit,label="profit")

plt.legend()
```

```python
#Histrogram chart

import matplotlib.pyplot as plt
%matplotlib inline

blood_sugar=[113,85,90,150,149,88,115,135,80,77,82,129]

plt.hist(blood_sugar,bins=4
         ,rwidth=0.9 ,color='r')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

bloodsugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
bloodsugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel("sugar range")
plt.ylabel("total no of patients")
plt.title('blood sugar analysis')


plt.hist([bloodsugar_men, bloodsugar_women],bins=[80,100,125,150],rwidth=0.95,colors=['g','b'],label=['men','women'])
plt.legend()
```

```python
#pie chart   # to find the percentage use autopct=%0.1%%

import matplotlib.pyplot as plt
%matplotlib inline

amount=[3500,600,300,450,1500]
name=["Rent","food","phone","car","other"]

plt.pie(amount,labels=name,radius=2,autopct="%0.1f%%" ,shadow=True, explode=[0,0.1,0,0,0])
plt.axis("equal")
plt.savefig("piechart.png")# taking picture
```
