```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import seaborn as sns
```

```python
#generate our dataset
#dataset=make_blobs(n_samples=200,centers=4,n_features=2,cluster_std=1.6,random_state=50)
```

```python
df = pd.read_csv("DATASET.csv")
df
```

```python
df['PhoneService'] = df['PhoneService'].fillna(0)
```

```python
df['gender'].replace(['Male','Female'],[0,1],inplace=True)
df['Partner'].replace(['Yes','No'],[1,0],inplace=True)
df['Dependents'].replace(['Yes','No'],[1,0],inplace=True)
df['PhoneService'].replace(['Yes','No'],[1,0],inplace=True)
df['MultipleLines'].replace(['No phone service','No','Yes'],[0,0,1],inplace=True)
df['OnlineSecurity'].replace(['No','Yes','No internet sevice'],[0,1,0],inplace=True)
df['DeviceProtection'].replace(['No','Yes','No internet sevice'],[0,1,0],inplace=True)
df['TechSupport'].replace(['No','Yes','No internet sevice'],[0,1,0],inplace=True)
df['StreamingTV'].replace(['No','Yes','No internet sevice'],[0,1,0],inplace=True)
df['StreamingMovies'].replace(['No','Yes','No internet sevice'],[0,1,0],inplace=True)
df['PaperlessBilling'].replace(['Yes','No'],[1,0],inplace=True)
df['InternetService'].replace(['DSL','Fiber optic','No'],[1,2,0],inplace=True)
df['Contract'].replace(['Month-to-month','One year','Two year'],[0,1,2],inplace=True)
df['PaymentMethod'].replace(['Electronic check','Mailed check','Bank transfer (automatic)','Credit card (automatic)'],[0,1,2,3],inplace=True)
df['Churn'].replace(['Yes','No'],[1,0],inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
total= le.fit_transform(df['TotalCharges'])
monthly = le.fit_transform(df['MonthlyCharges'])
df['TotalCharges'] = total
df['MonthlyCharges'] = monthly
df
f=df.drop(columns =['customerID','OnlineBackup','OnlineSecurity','DeviceProtection','TechSupport','StreamingTV','StreamingMovies'])
f.dtypes
f
```

```python
X = f.drop('TotalCharges',axis = 1)
y = f['TotalCharges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
print('Coefficients: \n', lm.coef_)
```

```python
predictions = lm.predict(X_test)
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
sns.distplot((y_test-predictions),bins=50)
```

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
predTree = drugTree.predict(X_test)
print (predTree [0:5])
print (y_test [0:5])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, predTree))
```

```python
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report
TeleTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
TeleTree.fit(X_train,y_train)
y_predtree = TeleTree.predict(X_test)
print(classification_report(y_test,y_predtree))
```
